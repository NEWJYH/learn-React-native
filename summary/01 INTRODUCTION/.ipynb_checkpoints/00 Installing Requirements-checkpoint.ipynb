- React Native 앱에서 가장 중요한 부분은
  
  Bridge들을 통해서 코드가 운영체제와 통신을 할 수 있도록 하는 인프라시설
  
  
- 기본시설은 JavaScript 코드가 운영체제와 이야기할 수 있도록 해줌

- Java가 설치되어 있어야 함

- .apk(안드로이드) .ipa(애플) 

- Java와 Xcode로 이 인프라를 가져와서 apk와 ipa 안에 넣어줌

- 이후 store에 보내짐

- React Native 앱은 이 모든 구조와 JavaScript 코드의 조합!!

- Expo app을 통하면 인프라시설이 이미 존재 하는 곳에 

  JavaScript / Markup/Styling만 넣으면 됨

Expo : https://expo.dev/

## 01 Expo 설치

```shell
npm install --global expo-cli
```

- compile을 해서 앱을 만드는게 아님
- Expo를 이용
- Expo는 React Native 코드를 곧바로 폰으로 전송해줌
- Expo는 폰 안에 있는 React Native 코드를 실행시킴
- 폰에서 React Native 코드를 바로 미리보기 할 수 있음
